```python
df['Review Rating']= df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(" ","_")
```

```python
df.columns
```

```python
df =  df.rename(columns = {'purchase_amount_(usd)':'purchase_amount'})
```

```python
df.columns
```

```python
#create a column age group
labels = ['Young Adult', 'Adult', 'Middle-aged', 'Senior']
df['age_group'] = pd.qcut(df['age'], q=4,labels = labels)
```

```python
df[['age','age_group']].head(20)
```

```python
df.head()
```

```python
#create a column purchase_frequency_days

frequency_mapping = {
    'Fortnightly' : 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly':90,
    'Annually': 365,
    'Every 3 months': 90
}

df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
import numpy as np
```

```python
df['purchase_frequency_days'] = df['purchase_frequency_days'].replace([np.inf, -np.inf], np.nan).fillna(0).round(0).astype(int)
```

```python
from sqlalchemy import create_engine

#step 1: Connect to postgree sql 
#Replace placeholders with your actual details
username = "postgres"   #default user
password = "postgrespassword"   #password set during installation
host = "localhost"    #if running locally
port = "5432"    #default postgre port
database = "customer_behaviour"  #database name created in pgAdmin

engine = create_engine(f"postgresql+psycopg2://{username}:{password}@{host}:{port}/{database}")

#step 2 : Load Dataframe into POstgreSQL
table_name = "customer"    #choose any table name
df.to_sql(table_name, engine, if_exists="replace",index = False)

print(f"Data Successfully loaded into table '{table_name}' in database '{database}'.")
```
